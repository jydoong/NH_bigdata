```python
from newspaper import Article

def extract_text(url):
    
    article = Article(url)
    article.download()
    article.parse()
    text = article.text or 'N/A'

    return text
```

```python
# Assuming url_list and df are already defined
df_copied = df.copy()
df_copied['text'] = None  # Initialize the 'text' column
```

```python
#몇번부터 시작할지(중간에 중단되었을 경우)
def save_checkpoints(nasdaq_df, start_index, end_index):
    #start_index = 20000
    empty_url_list = []

    for index, row in nasdaq_df.iterrows():
        if index < start_index:
            continue

        try:
            url = row['url_ifo']
            extracted_text = extract_text(url)
            
            if extracted_text:
                nasdaq_df.at[index, 'text'] = extracted_text
            else:
                print(f"Empty text extracted at index {index} for URL {url}")
                empty_url_list.append(url)

        except Exception as e:
            print(f"Error at index {index}: {e}")
        
        if index % 10 == 0:
            print(f"Processing index {index}")

        if index % 100 == 0:
            display(nasdaq_df.head())
        
        if index % 1000 == 0:
            nasdaq_df.to_csv(f"C:\\Users\\Administrator\\Bitamin_11\\dacon_nh\\dacon_venv\\dataset\\nasdaq_text_crawling_ckpt2\\{start_index}_{index}_ckpt.csv", index=False) #경로 변경
        
        if index == end_index:
            nasdaq_df.to_csv(f"C:\\Users\\Administrator\\Bitamin_11\\dacon_nh\\dacon_venv\\dataset\\nasdaq_text_crawling_ckpt2\\{start_index}_{index}_ckpt.csv", index=False) #경로 변경
            print(f"{start_index}_{index} 크롤링 완료")
            break
            
```

### ckpt 파일 불러와서 df로 합치기 - 여기까지는 완료

```python
#ckpt 불러와서 데이터프레임으로 합치기

import pandas as pd
import os
import re

PATH = "C:\\Users\\Administrator\\Bitamin_11\\dacon_nh\\dacon_venv\\dataset\\nasdaq_text_crawling_ckpt2\\"
csv_files = [f for f in os.listdir(PATH) if f.endswith('.csv')]

min_val, max_val = float('inf'), float('-inf')

# 각 CSV 파일을 읽어서 처리
for i, csv_file in enumerate(csv_files):
    # 데이터프레임 로드
    df = pd.read_csv(os.path.join(PATH, csv_file))

    if i == 0:
        result_df = df.copy()
    
    if i > 0:
        result_df[f'text_{i}'] = df['text']
    
    numbers = re.findall(r'\d+', csv_file)  # 숫자를 모두 찾음
    min_candidate, max_candidate = map(int, numbers[:2])  # 첫 두 개의 숫자를 사용 (0과 20000, 50000과 73000 등)

    # 최소값과 최대값 업데이트
    min_val = min(min_val, min_candidate)
    max_val = max(max_val, max_candidate)

print(f"Results ranging from {min_val} to {max_val}")
    
print()
result_df.head()
```

```python
#text_not_nan: 각 df의 text 열에서 nan이 아닌 값을 불러온 열

import numpy as np

# 'text'로 시작하는 모든 열을 필터링
text_columns = [col for col in result_df.columns if col.startswith('text')]

# 'NaN'이 아닌 첫 번째 값을 찾는 새로운 열 생성
result_df['text_not_nan'] = result_df[text_columns].apply(lambda row: next((item for item in row if not pd.isna(item)), np.nan), axis=1)

result_df
```

```python
from IPython.display import display

# 예시 nasdaq_df 데이터프레임과 nan_index_range 생성
#nasdaq_df = pd.DataFrame({'url_ifo': ['url1', 'url2', 'url3', 'url4'], 'text': [None, 'text2', None, 'text4']})
#nan_index_range = nasdaq_df[nasdaq_df['text'].isna()].index

# 함수 정의
def save_checkpoints(nasdaq_df, start_index=0, end_index=88769, nan_index_range=None): #원하는 index range에서 nan_index_range 찾을 수 있음
    empty_url_list = []
    error_url_list = []
    for index in nan_index_range:
        if index < start_index:
            continue

        if index > end_index:
            nasdaq_df.to_csv(f"C:\\Users\\Administrator\\Bitamin_11\\dacon_nh\\dacon_venv\\dataset\\nasdaq_text_crawling_ckpt2\\{start_index}_{index}_nan_indices_filled_ckpt.csv", index=False)
            print(f"{start_index}_{end_index} 크롤링 완료")
            break


        try:
            url = nasdaq_df.at[index, 'url_ifo']
            # 실제 크롤링 코드는 주석 처리 (여기서는 예시로 url을 그대로 사용)
            extracted_text = extract_text(url)
            
            if extracted_text:
                nasdaq_df.at[index, 'text'] = extracted_text
                print(f"Processing index {index}")
            else:
                print(f"Empty text extracted at index {index} for URL {url}")
                empty_url_list.append(url)

        except Exception as e:
            print(f"Error at index {index}: {e}")  
            error_url_list.append(url)     

    #error가 난 url이 있는 행은 빼고 저장
    nasdaq_df = nasdaq_df[~nasdaq_df['url_ifo'].isin(error_url_list)]
    nasdaq_df.to_csv("C:\\Users\\Administrator\\Bitamin_11\\dacon_nh\\dacon_venv\\dataset\\nan_indices_filled_{start_index}_{end_index}.csv", index=False)

    return nasdaq_df, empty_url_list, error_url_list
            

# 함수 호출 예시
#nasdaq_df, empty_url_list, error_url_list = save_checkpoints(nasdaq_df, 0, 3, nan_index_range)

#결과 확인
#nasdaq_df
```
