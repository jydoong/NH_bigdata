```python
# 필요한 라이브러리 import
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
news = pd.read_pickle('/Users/jiyoung/Desktop/대외활동/공모전/NH투자증권/data/all_matching_links_final.pkl')
```

```python
df = pd.DataFrame(news, columns = ['cnbc_url'])
df.head()
```

```python
# 크롬 드라이버 초기화
chrome_options = webdriver.ChromeOptions()
# chrome_options.add_argument('--headless')  # 크롬 브라우저를 헤드리스 모드로 실행 (화면 표시 안 함)
driver = webdriver.Chrome(options=chrome_options)

# 크롤링 결과를 저장할 리스트 초기화
article_categories = []

# cnbc 뉴스 기사 URL이 담긴 데이터프레임 df가 있다고 가정
for url in df['cnbc_url']:
    # 기사 페이지에 접근
    driver.get(url)
    
    # 기사 페이지의 HTML을 가져옴
    page_html = driver.page_source
    
    # BeautifulSoup을 사용하여 HTML 파싱
    soup = BeautifulSoup(page_html, 'html.parser')
    
    # 'ArticleHeader-eyebrow' 클래스를 가진 요소 추출
    article_header = soup.find(class_='ArticleHeader-eyebrow')
    
    if article_header:
        # 텍스트 내용을 추출하여 리스트에 추가
        article_categories.append(article_header.text)
    else:
        # 'ArticleHeader-styles-makeit-eyebrow--Degp4' 클래스를 가진 요소 추출
        article_header = soup.find(class_='ArticleHeader-styles-makeit-eyebrow--Degp4')
        if article_header:
            article_categories.append(article_header.text)
        else:
            # 두 클래스를 찾지 못한 경우에 대한 처리
            article_categories.append("N/A")

# 결과 출력
for category in article_categories:
    print(category)

# 크롬 드라이버 종료
driver.quit()
```

```python
df['article_categories'] = article_categories
df
```
