```python
company_list = df_list['tck_iem_cd'].unique()
```

```python
company_list
```

```python
relat = {}
for company in company_list:
    if company not in relat.keys():
        relat[company] = {}
        
    rel_com_list = df0922[df0922['tck_iem_cd'] == company]
    
    for lst in list(rel_com_list['all_tck_iem_cd']):
        lst = str(lst)
        lst = lst.replace('[', '')
        lst = lst.replace(']', '')
        lst = lst.replace("'", '')
        lst = lst.replace(" ", '')
        lst = lst.split(',')
        
        
        for l in lst:
            if l in relat[company].keys():
                relat[company][l] = relat[company][l]+1
            
            else:
                relat[company][l] = 1

relat
```

```python
df0922[df0922['tck_iem_cd'] == 'ANY']
```

```python
key_ls = list(relat.keys())  

for key in relat.keys():
    key_l = list(relat[key].keys())
    for k in key_l:
        if key == k:
            del relat[key][k]

for key in key_ls:
    if not relat[key]:
        del relat[key] 


compant_relat_data = {}
for key in relat.keys():
    relat[key] = dict(sorted(relat[key].items(), key=lambda x:x[1], reverse=True))
    cnt = 0
    compant_relat_data[key] = {}
    for k in relat[key].keys():
        compant_relat_data[key][k] = relat[key][k]
        cnt += 1
        if cnt == 5:
            break
compant_relat_data
```

```python
compant_relat_data_v2 = {}
for key in compant_relat_data.keys():
    for j in compant_relat_data[key].keys():
        if compant_relat_data[key][j] >= 6:
            compant_relat_data_v2[key] = compant_relat_data[key]
        break
len(compant_relat_data_v2.keys())
compant_relat_data_v2['TSLA']
```
