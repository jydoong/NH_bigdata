```python
# 각 분야에 해당하는 회사 dictionary
cp_from_theme = {}

for i in range(len(df_na_drop)):
    lst = df_na_drop.iloc[i]['description']
    lst = lst.lower()
    lst = lst.replace(',', '')
    lst = lst.replace('.', '')
    lst = lst.split(" ")
    for word in key_dic.keys():
        for wrd in key_dic[word]:
            if wrd in lst:
                if word in cp_from_theme.keys():
                    cp_from_theme[word].append(df_na_drop.iloc[i]['tck_iem_cd'])
                    
                else:
                    cp_from_theme[word] = [df_na_drop.iloc[i]['tck_iem_cd']]

for k in key_dic.keys():
    cp_from_theme[k] = set(cp_from_theme[k])

cp_from_theme
pickle.dump(cp_from_theme, open('/Users/parkjeonghoon/Desktop/BITamin/NH/cp_from_theme.pkl', 'wb'))
```
