### stock code 입력하여 nasdaq_df에서 해당 기사의 finbert 결과 찾기

```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
#FinBERT 모델 다운로드
tokenizer = AutoTokenizer.from_pretrained("ProsusAI/finbert")
model = AutoModelForSequenceClassification.from_pretrained("ProsusAI/finbert")
```

```python
#해결방법: 여기서는 놔두고, 나중에 추천 기업 리스트 뜨면 stock_info.csv에서 재무정보가 N/A로 뜨는 기업 거르는 거 어떨지?
```

```python
def find_rows_by_exact_tck(df, stock_code):
    rows_to_append = []  # List to store rows that match the stock_code
    
    for i in range(len(df)):
        lst = df['all_tck_iem_cd'].iloc[i]
        lst = lst.replace('[', '')
        lst = lst.replace(']', '')
        lst = lst.replace("'", '')
        lst = lst.replace(" ", '')
        lst = lst.split(',')
        
        if stock_code in lst:
            a = df.iloc[i]
            rows_to_append.append(a)
    
    # Concatenate all the rows at once
    new_df = pd.concat(rows_to_append, axis=1).T.reset_index(drop=True)
    return new_df

#주어진 데이터프레임의 'text' 열에서 stock_code, company_name를 포함한 문단을 추출
def get_related_paragraphs(row, stock_code, company_name): 
    # Use the 'text' field from the passed row
    text = row['text']
    
    # Split the article into paragraphs
    paragraphs = text.split('\n\n')
    
    # Filter paragraphs containing the stock_code or company_name
    relevant_paragraphs = [paragraph for paragraph in paragraphs if stock_code in paragraph or company_name in paragraph]
    
    # Join relevant paragraphs into a single string
    result = '\n\n'.join(relevant_paragraphs)
    return result


#추출한 문단에 FinBERT 적용
def finbert(input_text): 
    inputs = tokenizer(input_text, padding = True, truncation = True, return_tensors='pt')
    outputs = model(**inputs)
    predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
    positive = predictions[:, 0].tolist()
    negative = predictions[:, 1].tolist()
    neutral = predictions[:, 2].tolist()

    table = {
            "Text": input_text,
            "Positive":positive,
            "Negative":negative,
            "Neutral":neutral}

    df = pd.DataFrame(table, columns = ["Text", "Positive", "Negative", "Neutral"])

    return df

#nasdaq_df, stock_code를 입력하면 FinBERT 결과 반환
def finbert_from_stock_code(df, stock_code): 
    dfs = pd.DataFrame()
    new_df = find_rows_by_exact_tck(df, stock_code)
    
    # Assuming stock_info_df is globally defined and contains 'fc_sec_eng_nm'
    company_name = stock_info_df[stock_info_df['tck_iem_cd'] == stock_code]['fc_sec_eng_nm'].iloc[0]
    
    for index, row in new_df.iterrows():
        result = get_related_paragraphs(row, stock_code, company_name)  # Passing the entire row
        df = finbert(result)
        dfs = pd.concat([dfs, df], ignore_index=True)
        dfs = dfs[dfs['Text'].str.len() > 0]

    return dfs
```

```python
#각 기사에 대한 FinBERT 결과

result_df = finbert_from_stock_code(nasdaq_df.iloc[:10000], 'AAPL')
result_df
```
